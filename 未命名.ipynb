```python
from sklearn.externals import joblib 
import numpy as np 
from sklearn.linear_model import SGDRegressor


metadata = np.load('./10_australian_big_metadata.npy')
X = metadata[:, 0:396]
y = metadata[:, 396]

auto_regressor = joblib.load('./automl.joblib')
print(auto_regressor.show_models())
```

```python
print(auto_regressor.cv_results_)
```

```python
auto_regressor.get_models_with_weights()
```

```python
auto_regressor.get_params()
```

```python
print(auto_regressor.sprint_statistics())
```

```python
pred = auto_regressor.predict(X)
from sklearn.metrics import r2_score

print("R2 score : ", r2_score(y, pred))
```

```python
sgd_r = SGDRegressor(learning_rate='invscaling', alpha=0.0001, eta0=0.005, loss='huber')
sgd_r.fit(X, y)
sgd_pred = sgd_r.predict(X)
print("R2 score : ", r2_score(y, sgd_pred))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
lr_pred = lr.predict(X)
print("R2 score : ", r2_score(y, lr_pred))
```

```python
import copy
from sklearn.metrics import accuracy_score
auto_classifier = joblib.load('./autoclassifier.joblib')

y_train = copy.deepcopy(y)
y_train[np.where(y_train>0)[0]] = 1
y_train[np.where(y_train<=0)[0]] = -1

c_pred = auto_classifier.predict(X)
print("accurcay score : ", accuracy_score(y_train, c_pred))
```

```python
from sklearn.linear_model import LogisticRegression

clr = LogisticRegression()
clr.fit(X, y_train)
clr_pred = clr.predict(X)
print("accurcay score : ", accuracy_score(y_train, clr_pred))
```
